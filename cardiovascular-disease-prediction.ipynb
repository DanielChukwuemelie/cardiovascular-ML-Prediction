***The Dataset***

The CVD dataset for this project contains the following Features :

●id :> It's just the Id no of the row. Not relevant.

●age :> It's the age of a person in Days.

●gender :> It's the gender of the person.

●height :> It's the height of the person in cm.

●weight :> It's the weight of the person in kg.

●ap_hi :> It's the Systolic blood pressure i.e. Pressure exerted when Blood is ejected in arteries. Normal value : 120mmhg or Below.

●ap_low :> It's the Diastolic blood pressure i.e. Pressure exerted when Blood exerts between arteries and heartbeats. Normal Value : 80mmhg or Below.

●cholesterol :> It's the Cholesterol value (Cholesterol is a type of fat found in your blood) of your blood. In Adults, 200 mg/dL is desired with 200 and 239 mg/dL as Boderline High. In Children, 170 mg/dL is desired with 170 and 199 mg/dL as Boderline High.

●gluc :> It's the Glucose Level. They're less than 100 mg/dL after not eating (fasting) for at least 8 hours. And they're less than 140 mg/dL 2 hours after eating. For most people without diabetes, blood sugar levels before meals hover around 70 to 80 mg/dL.

●smoke :> It contains Binary Values stating whether Person is a Smoker or not i.e. {0 : 'Not a Smoker', 1 : 'Smoker'}.

●alco :> It contains Binary Values stating whether a Person is an alcoholic or not i.e. {0 : 'Not a Alcoholic', 1 : 'Alcoholic'}.

●active :> It contains Binary Values stating whether a Person is involved in physical activities or not i.e. {0 : 'Not involved in Physical Activities', 1 : 'involved in physical activities'}.

●cardio :> It's our Target Value Binary Values stating whether a Person has Cardiovascular diseases (CVDs) or Not i.e. {0 : 'Not Have CVD', 1 : 'Have CVD'}.

# Performance Metrics

* ROC curve and AUC score: An ROC curve (receiver operating characteristic curve) is a graph showing the performance of a                              classification model at all classification thresholds. While an AUC (Area Under the Curve)                                  quantifies the overall performance of the model across all possible thresholds.

* Confusion matrix: A confusion matrix is a summary of prediction results on a classification problem. The number of                           correct and incorrect predictions are summarized with count values and broken down by each class. This                     is the key to the confusion matrix.

* Classification report: A classification report is a performance evaluation metric in machine learning. It is used to show                          the precision, recall, F1 Score, and support of your trained classification model.

* Precision: Precision is a metric that measures how often a machine learning model correctly predicts the positive class.              It is calculated as the number of true positives divided by the sum of true positives and false positives.

* Recall: Recall is a metric that measures the percentage of data samples that a machine learning model correctly                     identifies as belonging to a class of interest—the “positive class”—out of the total samples for that class. It             is also known as the true positive rate (TPR).

* F1-score: The F1 score can be interpreted as a harmonic mean of the precision and recall, where an F1 score reaches its               best value at 1 and worst score at 0.

We are agoing to try 3 Machine learning models:

1. Logistic Regression
2. XGBoost Classifier
3. Random Forest Classifier

```python
# Put models in a dictionary, this helps to avoid going over all the models one by one in fitting and all.
models = {"Logistic Regression": LogisticRegression(),
          "XGB": XGBClassifier(),
          "Random Forest": RandomForestClassifier()}

# Create a function to fit and score models
def fit_and_score(models, X_train, X_test, y_train, y_test):
    """
    Fits and evaluates machine learning models.
    models : a dictionary of different Scikit-learn machine learning models
    X_train : training data (no labels)
    X_test : testing data (no labels)
    y_train : training labels
    y_test : testing labels
    """
    # Set random seed
    np.random.seed(42)
    # Make a dictionary to keep model scores
    model_scores = {}
    # Loop through models
    for name, model in models.items():
        # Fit the model to the data
        model.fit(X_train, y_train)
        # Evaluate the model and append its score to model_scores
        model_scores[name] = model.score(X_test, y_test)
    return model_scores
```

```python
model_scores = fit_and_score(models=models,
                             X_train=X_train,
                             X_test=X_test,
                             y_train=y_train,
                             y_test=y_test)
model_scores
```

# Hyperparameter Tuning By Hand

```python
# Let's tune LogisticRegression

# Define the logistic regression model
model = LogisticRegression()

# Define lists of different values for hyperparameters
Cs = [0.1, 1.0, 10.0]
penalties = ['l1', 'l2']
solvers = ['liblinear', 'saga']
max_iters = [100, 200, 300]

train_scores = []
test_scores = []

# Loop through different values of hyperparameters
for C in Cs:
    for penalty in penalties:
        for solver in solvers:
            for max_iter in max_iters:
                # Set the hyperparameter values for the logistic regression model
                model.C = C
                model.penalty = penalty
                model.solver = solver
                model.max_iter = max_iter

                # Fit the model on the training data
                model.fit(X_train, y_train)

                # Calculate the training and testing accuracy scores
                train_score = model.score(X_train, y_train)
                test_score = model.score(X_test, y_test)

                # Append the scores to the lists
                train_scores.append(train_score)
                test_scores.append(test_score)
```

```python
train_scores
```

```python
test_scores
```

# Lets visualize it


```python
# Create empty arrays to store the scores
train_scores = np.zeros((len(max_iters), len(penalties), len(Cs), len(solvers)))
test_scores = np.zeros((len(max_iters), len(penalties), len(Cs), len(solvers)))

# Loop through hyperparameters
for i, max_iter in enumerate(max_iters):
    for j, penalty in enumerate(penalties):
        for k, C in enumerate(Cs):
            for l, solver in enumerate(solvers):
                # Create an instance of the logistic regression model
                model = LogisticRegression(C=C, penalty=penalty, solver=solver, max_iter=max_iter)

                # Fit the model on the training data
                model.fit(X_train, y_train)

                # Calculate the training and testing accuracy scores
                train_score = model.score(X_train, y_train)
                test_score = model.score(X_test, y_test)
                
                # Store the scores in the arrays
                train_scores[i, j, k, l] = train_score
                test_scores[i, j, k, l] = test_score

# Create a heatmap for a subset of the test scores
fig, ax = plt.subplots(figsize=(10, 7))
sns.heatmap(test_scores.mean(axis=(1, 3)), annot=True, fmt=".3f", cmap="Blues", xticklabels=solvers, yticklabels=penalties)
ax.set_xlabel("Solver")
ax.set_ylabel("Penalty")
ax.set_xticklabels(solvers, rotation=45)
ax.set_yticklabels(penalties, rotation=0)
ax.set_title("Logistic Regression Test Scores Heatmap")

# Show the plot
plt.show()
```

```python
max_score = np.max(test_scores)
print(f"Maximum Logistic Regression score on the test data: {max_score * 100:.2f}%")
```

# Hyperparameter Tuning with RandomizedSearchCV

We are going to tune:

1. XGBClassifier()

...using RandomizedSearchCV

```python
# Create a hyperparameter grid for XGBClassifier
xgb_grid ={
    'learning_rate': [0.01, 0.001],
    'max_depth': [3, 5, 7],
    'n_estimators': [100, 200, 300],
    'subsample': [0.8, 0.9, 1.0],
    'colsample_bytree': [0.8, 0.9, 1.0]
}

# Create hyperparameter grid for RandomForestClassifier
rf_grid = {"n_estimators": np.arange(10, 1000, 50),
           "max_depth": [None, 3, 5, 10],
           "min_samples_split": np.arange(2, 20, 2),
           "min_samples_leaf": np.arange(1, 20, 2)}
```

Now we have got hyperparameters grids setup for each of our models, lets tune them using RandomizedSearcCV

```python
# Lets tune our random forest

# Setup random seed
np.random.seed(42)

# Setup random hyperparameter search for XGBClassifier
rs_xg = RandomizedSearchCV(xgb.XGBClassifier(),
                           param_distributions=xgb_grid,
                           cv=5,
                           n_iter=20,
                           refit=True, # set refit=True (default) to refit the best model on the full dataset
                           verbose=True)

# Fit random hyperparameter search model for XGBClassifier()
rs_xg.fit(X_train, y_train)
```

```python
rs_xg.best_params_
```

```python
rs_xg.score(X_test, y_test)
```

```python
# Setup random seed
np.random.seed(42)

# Setup random hyperparameter search for RandomSearchClassifier
rs_rf = RandomizedSearchCV(RandomForestClassifier(),
                           param_distributions=rf_grid,
                           cv=5,
                           n_iter=20,
                           refit=True, # set refit=True (default) to refit the best model on the full dataset
                           verbose=True)

# Fit random hyperparameter search model for RandomForestClassifier()
rs_rf.fit(X_train, y_train)
```

```python
# Find the best hyperparamters
rs_rf.best_params_
```

```python
# Evaluate the randomized search RandomForestClassifier model
rs_rf.score(X_test, y_test)
```

# Calculate evaluation metrics using cross-validation

We are going to calculate the accuracy, precision, recall and f1-score of our model using cross-validation and to do so we will be using "cross_val_score()".

```python
# Check best hyperparameters
rs_xg.best_params_
```

```python
# Create a new classifier with best parameters
clf = XGBClassifier(subsample = 0.8,
                    n_estimators = 300,
                    max_depth = 7,
                    learning_rate = 0.01,
                    colsample_bytree = 0.8)
```

```python
# Cross-validataed Accuracy
cv_acc = cross_val_score(clf,
                         X,
                         y,
                         cv=5,
                         scoring="accuracy")

cv_acc = np.mean(cv_acc)
cv_acc
```

```python
# Cross-validataed Precision
cv_precision = cross_val_score(clf,
                         X,
                         y,
                         cv=5,
                         scoring="precision")
cv_precision = np.mean(cv_precision)
cv_precision
```

```python
# Cross-validataed Recall
cv_recall = cross_val_score(clf,
                         X,
                         y,
                         cv=5,
                         scoring="recall")
cv_recall = np.mean(cv_recall)
cv_recall
```

```python
# Cross-validataed F1-Score
cv_f1 = cross_val_score(clf,
                         X,
                         y,
                         cv=5,
                         scoring="f1")
cv_f1 = np.mean(cv_f1)
cv_f1
```

```python
# Visualize cross validation metrics
cv_metrics = pd.DataFrame({"Accuracy": cv_acc,
                           "Precision": cv_precision,
                           "Recall": cv_recall,
                           "F1": cv_f1},
                           index=[0])

cv_metrics.T.plot.bar(title="Cross-Validated Classification Metrics",
                      legend=False);
```
